```python
def plan_data(data,plan='All'):
    data2=data.copy()
    if plan=='All':
        data2=data2
    elif plan=='Annual':
        data2=data2[data2['subscription_type']=='Annual']
    elif plan=='Monthly':
        data2=data2[data2['subscription_type']=='Monthly']
    elif plan=='Quarterly':
        data2=data2[data2['subscription_type']=='Quarterly']
    data2=data2.reset_index()
    return(data2)
```

# Pre-Processing Data

```python
def remove_page_duplicates(data,target_column):
    data2=data.copy()
    user_journey=list(data2[target_column])
    lst_user_journey_split=[]
    for row in user_journey:
        lst_user_journey_split.append(row.split('-'))
    
    l1=[]
    for i in lst_user_journey_split:
        l2=[]
        for j in i:
            if j not in l2:
                l2.append(j)
        l1.append(l2)

    output=[]
    for i in l1:
        res='-'.join([str(j) for j in i])
        output.append(res)
    data2[target_column]=pd.Series(output)
    
    return(data2)
```

```python
def group_by(data,group_column,target_column,count_from,sessions='All'):
    data2=data.copy()
    user_id=data2[group_column].unique()
    
    
    o=[]
    if sessions=='All':
        for i in user_id:
            u_j='-'.join([str(data2[target_column][j])  for j in data2.index  if i==data2[group_column][j]])
            o.append(u_j)
    else:
        if count_from=='first':
            t=[]
            for i in user_id:
                t1=[]
                for j in data2.index:
                    if i==data2[group_column][j]:
                        t1.append([data2[target_column][j]])
                t1=t1[:sessions]
                t2=[]
                for i in t1:
                    for j in i:
                        t2.append(str(j))
                u_j='-'.join(t2)
                o.append(u_j)            
        elif count_from=='last':
            t=[]
            for i in user_id:
                t1=[]
                for j in data2.index:
                    if i==data2[group_column][j]:
                        t1.append([data2[target_column][j]])
                t1=t1[-sessions:]
                t2=[]
                for i in t1:
                    for j in i:
                        t2.append(str(j))
                u_j='-'.join(t2)
                o.append(u_j)
    data3=pd.DataFrame(list(zip(user_id,o)),columns=['user_id','user_journey'])
    return(data3)
```

```python
def remove_pages(data,pages,target_column):
    data2=data.copy()
    user_id=data2['user_id']
    user_journey=list(data2[target_column])
    
    t=[]
    for i in user_journey:
        x=i.split('-')
        for j in pages:
            if j in x:
                x.remove(j)
        t.append(x)
    output=[]
    for i in t:
        res='-'.join([str(j) for j in i])
        output.append(res)
    data3=pd.DataFrame(list(zip(user_id,output)),columns=['user_id','user_journey'])
    return(data3)
```

```python
def Page_count(data,target_column,plan='All'):
    data2=plan_data(data,plan)
    x=data2[target_column]
    
    l=[]
    for i in x:
        j=i.split('-')
        l.append(j)
    
    m=[]
    for i in l:
        for j in i:
            if j not in m:
                m.append(j)
    n=[]
    for i in m:
        count=0
        for j in l:
            for k in j:
                if i==k:
                    count+=1
        n.append([i,count])
    df=pd.DataFrame(data=n,columns=['Page','Page Count'])
    return(df)
```

```python
def Page_presence(data,target_column,plan='All'):
    data2=plan_data(data,plan)
    data2=remove_page_duplicates(data2,target_column)
    x=data2[target_column]

        
    # print(x.isna().sum())
    l=[]
    for i in x:
        if type(i) is not str:
            pass
        else:
            j=i.split('-')
            l.append(j)
    
    m=[]
    for i in l:
        for j in i:
            if j not in m:
                m.append(j)
    n=[]
    for i in m:
        count=0
        for j in l:
            for k in j:
                if i==k:
                    count+=1
        n.append([i,count])
    df=pd.DataFrame(data=n,columns=['Page','Page Count'])
    return(df)
```

```python
def Page_destination(data,target_column,plan='All'):
    import numpy as np
    data2=plan_data(data,plan)
    # data2=remove_page_duplicates(data2,target_column)
    x=data2[target_column]

    l=[]
    for i in x:
        j=i.split('-')
        l.append(j)
    m=[]
    for i in l:
        for j in i:
            if j not in m:
                m.append(j)
    df=pd.DataFrame(columns=m,index=m)

    for i in m:
        c=[]
        for j in l:
            for k in range(len(j)):
                if i==j[k]:
                    try:
                        c.append(j[k+1])
                    except:
                        pass
        c1=set(c)
        c2={}
        for o in c1:
            count=0
            for p in c:
                if o==p:
                    count+=1
            c2[o]=count
        for k,v in c2.items():
            df[i][k]=v
    df.fillna(0, inplace=True)
    print('Read as: Page in Column is followed by Page in index by x number of times') 
    return(df)
```

```python
def Page_sequence(data,target_column,page_sequence,plan='All'):
    data2=plan_data(data,plan)
    data2=remove_page_duplicates(data2,target_column)
    x=data2[target_column]
    page_sequence='-'.join(page_sequence)
    count=0
    for i in x:
        count=count+i.count(page_sequence)
    return(count)
```

```python
def Journey_length(data,target_column,plan='All'):
    data2=plan_data(data,plan)
    data2=remove_page_duplicates(data2,target_column)
    x=data2[target_column]

    x1=[]
    for i in x:
        x1.append(i.split('-'))
    length=0
    count=0    
    for i in x1:
        length+=len(i)
        count+=1
    import numpy as np
    j_l=np.round(length/count,2)
    return(str(j_l)+' Pages')
```
