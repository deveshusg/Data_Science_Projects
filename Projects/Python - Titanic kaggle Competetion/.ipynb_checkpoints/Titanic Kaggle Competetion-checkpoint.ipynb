```python
import pandas as pd
import numpy as np
train=pd.read_csv('train.csv')
train.head()
```

```python
np.round(100*train.isnull().sum()/len(train),2)
```

```python
train=train.drop(['Cabin','PassengerId'],axis=1)
```

```python
train=train.dropna()
```

```python
np.round(100*train.isnull().sum()/len(train),2)
```

```python
train.describe(include='all')
```

```python
y=train['Survived']
x1=train.drop(['Survived','Name','Ticket','Fare'],axis=1)
```

```python
x1['Sex']=x1['Sex'].map({'male':0,'female':1})
x1['Embarked']=x1['Embarked'].map({'S':0,'C':1,'Q':2})
```

```python
x1
```

```python
import statsmodels.api as sm
```

```python
x=sm.add_constant(x1)
```

```python
reg=sm.OLS(y,x)
```

```python
reg.fit()
```
